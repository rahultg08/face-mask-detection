```python
import os
```

```python
from keras.preprocessing import image
```

```python
import cv2
```

```python
categories = ['with_mask','without_mask']
```

```python
data=[]
for category in categories:
    path=os.path.join('train',category)
    label=categories.index(category)  #used for indexing, 0 for with mask and 1 for without mask
    
    for file in os.listdir(path):
        
        img_path = os.path.join(path,file)
        img = cv2.imread(img_path)
        img = cv2.resize(img,(224,224))
        
        data.append([img,label])
```

```python
import random
```

```python
random.shuffle(data)            #used for shuffling the data as images are in the order with_mask -> without_mask
```

```python
X=[]
y=[]

for features,label in data:
    X.append(features)
    y.append(label)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from keras.applications.vgg16 import VGG16
```

```python
vgg = VGG16()
```

```python
import tensorflow as tf
```

```python
print(tf.__version__)
```

```python
vgg.summary()
```

```python
from keras import Sequential
```

```python
model= Sequential()
```

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.summary()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
#Adam optimization is a stochastic gradient descent method that is based on adaptive estimation of first-order and second-order moments
#It is a replacement optimization algorithm for stochastic gradient descent for training deep learning models
```

```python
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
def detect_facemask(img):
    y_pred = (model.predict(img.reshape(1,224,224,3))>0.5).astype(np.int32)
    return y_pred[0][0]
```

```python
def draw_label(img,text,pos,bg_color):
    text_size = cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)
    end_x = pos[0] + text_size[0][0] + 2
    end_y = pos[1] + text_size[0][1] - 2
    
    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1,cv2.LINE_AA)
```
